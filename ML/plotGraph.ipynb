```python
import pandas as pd
import mplfinance as mpf
from datetime import datetime, timedelta
import plotly
import plotly.graph_objs as go
from plotly.subplots import make_subplots
```

```python
df1h = pd.read_csv("stock_data1H.csv")
df1h = df1h[(df1h.Symbol == 'KBANK')].tail(100)
df1h['Datetime'] = pd.to_datetime(df1h['Datetime'])
my_range = pd.date_range(start= min(df1h['Datetime']), end= max(df1h['Datetime']), freq='H')
missing_datetime = my_range.difference(df1h['Datetime']).strftime("%Y-%m-%d H%:m%:s%").tolist()
```

```python
fig = go.Figure(data=[go.Candlestick(
                        x=df1h['Datetime'],
                        open=df1h['Open'],
                        high=df1h['High'],
                        low=df1h['Low'],
                        close=df1h['Close']
)])

fig.update_xaxes(
    rangebreaks=[
        dict(bounds=[17, 10], pattern="hour"),
        dict(values=missing_datetime, dvalue=3600000)
    ])

fig.show()
```

```python
df1d = pd.read_csv("stock_data1D.csv")
df1d = df1d[(df1d.Symbol == 'KBANK')].tail(100)
df1d['Date'] = pd.to_datetime(df1d['Date'])
my_range = pd.date_range(start= min(df1d['Date']), end= max(df1d['Date']), freq='B')
missing_date = my_range.difference(df1d['Date']).strftime("%Y-%m-%d").tolist()
df1d
```

```python
fig = go.Figure(data=[go.Candlestick(
                        x=df1d['Date'],
                        open=df1d['Open'],
                        high=df1d['High'],
                        low=df1d['Low'],
                        close=df1d['Close']
)])

fig.update_xaxes(
    rangebreaks=[
        dict(bounds=["sat", "mon"]),
        dict(values=missing_date) 
])

fig.show()
```

```python
agg_dict = {'Open': 'first',
          'High': 'max',
          'Low': 'min',
          'Close': 'last',
          'Adj Close': 'last',
          'Volume': 'sum'}

df1d = pd.read_csv("stock_data1D.csv")
df1d['Date'] = pd.to_datetime(df1d['Date'])
df1d.set_index('Date', inplace=True)
df1d = df1d[(df1d.Symbol == 'KBANK')].tail(100)
re_df1w = df1d.resample('W').agg(agg_dict)
re_df1w = re_df1w.reset_index()
```

```python
fig = go.Figure(data=[go.Candlestick(
                        x=re_df1w['Date'],
                        open=re_df1w['Open'],
                        high=re_df1w['High'],
                        low=re_df1w['Low'],
                        close=re_df1w['Close']
)])

fig.show()
```

```python
df1d = pd.read_csv("stock_data1D.csv")
df1d['Date'] = pd.to_datetime(df1d['Date'])
df1d.set_index('Date', inplace=True)
df1d = df1d[(df1d.Symbol == 'KBANK')].tail(100)
re_df1d = df1d.resample('D').agg(agg_dict)
# re_df1w = re_df1w.reset_index()
re_df1d
```

```python
new_df1d = re_df1d.dropna()
new_df1d_index = new_df1d.index.strftime("%Y.%m.%d")
# %H:%M:%S
new_df1d_index
```

```python
fig = make_subplots(rows=2, cols=1, shared_xaxes=True, 
               vertical_spacing=0.1, subplot_titles=('Candlestick', 'Volume'), 
               row_width=[0.5, 1])

fig.add_trace(go.Candlestick(
                        x=new_df1d_index,
                        open=new_df1d['Open'],
                        high=new_df1d['High'],
                        low=new_df1d['Low'],
                        close=new_df1d['Close']
),row=1, col=1)

fig.add_trace(go.Bar(x=new_df1d_index, y=new_df1d['Volume'], showlegend=False), row=2, col=1)

fig.update_xaxes(showticklabels=False)
fig.update(layout_xaxis_rangeslider_visible=False)
fig.show()
```

```python
fig_test = make_subplots(rows=2, cols=1, shared_xaxes=True, 
               vertical_spacing=0.08, subplot_titles=('Candlestick', 'Volume'))

trace1 = go.Candlestick(x=re_df1w['Date'], open=re_df1w['Open'],
                        high=re_df1w['High'], low=re_df1w['Low'],
                        close=re_df1w['Close'], visible=False, name='1 Week bar')

trace2 = go.Bar(x=re_df1w['Date'], y=re_df1w['Volume'], showlegend=False)

trace3 = go.Candlestick(x=re_df1m['Date'], open=re_df1m['Open'],
                        high=re_df1m['High'], low=re_df1m['Low'],
                        close=re_df1m['Close'], visible=False, name='1 Month bar')

trace4 = go.Bar(x=re_df1m['Date'], y=re_df1m['Volume'], showlegend=False)

trace5 = go.Candlestick(x=new_df1d_index, open=new_df1d['Open'],
                        high=new_df1d['High'], low=new_df1d['Low'],
                        close=new_df1d['Close'], visible=True, name='1 Day bar')

trace6 = go.Bar(x=new_df1d_index, y=new_df1d['Volume'], showlegend=False)

trace7 = go.Candlestick(x=df1h['Datetime'], open=df1h['Open'],
                        high=df1h['High'], low=df1h['Low'],
                        close=df1h['Close'], visible=False, name='1 Hour bar')

trace8 = go.Bar(x=df1h['Datetime'], y=df1h['Volume'], showlegend=False)

fig_test.append_trace(trace1,1,1)
fig_test.append_trace(trace2,2,1)
fig_test.append_trace(trace3,1,1)
fig_test.append_trace(trace4,2,1)
fig_test.append_trace(trace5,1,1)
fig_test.append_trace(trace6,2,1)
fig_test.append_trace(trace7,1,1)
fig_test.append_trace(trace8,2,1)
  
fig_test.update_layout(height=700,updatemenus=[
                    dict(type='buttons',
                        buttons=[
                        dict(label='1 Hour', method='update',
                            args=[{'visible': [False, False, False, False, False, False, True, True]}, {'title': '1 Hour'}]),
                        dict(label='1 Day', method='update',
                            args=[{'visible': [False, False, False, False, True, True, False, False]}, {'title': '1 Day'}]),
                        dict(label='1 Week', method='update',
                            args=[{'visible': [True, True, False, False, False, False, False, False]}, {'title': '1 Week'}]),
                        dict(label='1 Month', method='update',
                            args=[{'visible': [False, False, True, True, False, False, False, False]}, {'title': '1 Month'}])
                        ])])

fig_test.update_xaxes(showticklabels=False)
fig_test.update(layout_xaxis_rangeslider_visible=False)
fig_test.show()
```
