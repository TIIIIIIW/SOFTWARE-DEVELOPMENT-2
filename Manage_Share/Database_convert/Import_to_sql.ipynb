# Change Column Symbol in table satistics,financial,stock,sector,info

```python
# สร้าง เงื่อนไข ในการสร้าง column SymbolId 
Symbolid = table_info["SymbolId"].tolist()
Symbol =  table_info["Symbol"].tolist()
dic = {}
for i in range(len(Symbol)):
    
    dic[Symbol[i]] = Symbolid[i]
    
print(dic)
```

```python
# สร้าง เงื่อนไข ในการสร้าง column IndustryId
Industryid = table_industry["IndustryId"].tolist()
Industry =  table_industry["Full_Industry"].tolist()
dic_Industry = {}
for i in range(len(Industry)):
    
    dic_Industry[Industry[i]] = Industryid[i]
    
print(dic_Industry)
```

```python
# สร้าง เงื่อนไข ในการสร้าง column SectorId
Sectorid = table_sector["SectorId"].tolist()
Sector =  table_sector["Full_Sector"].tolist()
dic_Sector = {}
for i in range(len(Sector)):
    
    dic_Sector[Sector[i]] = Sectorid[i]
    
print(dic_Sector)
```
