```python
def save_csv(df,name):

    name += ".csv"
    df.to_csv(name,index=False)
    
    return "succes"
```

# INDUSTRY

```python
Full_Industry = []
short_Industry = []

for stock in data_df.index :

    
    
    if str(stock) != 'nan':
        
        target = stock.split('-')
        
        if target[0][:-1] in Industry:
            
            short_Industry.append(target[0][:-1])
            Full_Industry.append(target[1][1:])
```

```python
d = {'Short_Industry':short_Industry,'Full_Industry':Full_Industry}
```

```python
Industry_table = pd.DataFrame(data=d)
```

```python
Industry_table["IndustryId"] = Industry_table.index + 1
```

```python
Industry_table
```

```python
save_csv(Industry_table,"Industry")
```

```python
def get_share(data_df,all_share):
    
    Section_Industry = ['AGRO', 'AGRI', 'FOOD', 'CONSUMP', 'FASHION', 'HOME', 'PERSON', 'FINCIAL', 'BANK', 'FIN', 'INSUR', 'INDUS', 'AUTO', 'IMM', 'PAPER', 'PETRO', 'PKG', 'STEEL', 'PROPCON', 'CONMAT', 'PROP', 'PF&REIT', 'CONS', 'RESOURC', 'ENERG', 'MINE', 'SERVICE', 'COMM', 'HELTH', 'MEDIA', 'PROF', 'TOURISM', 'TRANS', 'TECH', 'ETRON', 'ICT']
    
    for i in data_df['']:

        if str(i) != 'nan' and not(str(i.split()[0]) in all_Industry_Sector):
            print(i.split()[0])
            all_share.append(i.split()[0])
            
    return all_share
```

```python
def get_SET_share(driver):
    
    Industrys = ['AGRO','CONSUMP','INDUS','PROPCON','RESOURC','SERVICE','TECH']
    all_share = []
    
    for Industry in Industrys:
        
        driver.get("https://www.set.or.th/th/market/index/set/"+Industry)
        driver.execute_script("scrollBy(0,+5000);")
        time.sleep(2)
        
        data = driver.page_source
        data_df = pd.read_html(data)[1]
        
        all_share = get_share(data_df,all_share)
        
    return all_share
```
