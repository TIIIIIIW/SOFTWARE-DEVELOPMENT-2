```python
def Separate_period_new_period(row):

    print(row['index'])
    return row['index'][:8]
    
def Separate_period_new_Date(row):

    return row['index'][8:]
    
def get_set50_100_df(driver):
    
    data = driver.page_source
    data_df = pd.read_html(data)[2]
    clean_column = []
    for column in data_df.columns:

        clean_column.append(column.replace("  (Click to sort Ascending)",""))
        
    data_df.columns = clean_column
    data_df = data_df.set_index('Symbol')
    return data_df

def operation_data(stock,dataframe):
    
    Financial_df = (dataframe.set_index("Period as of")).T
    Financial_df = Financial_df.drop(["Financial Data","Financial Ratio"], axis=1)

    Financial_df['Symbol'] = stock
    Financial_df = Financial_df.reset_index()
    Financial_df['Period as of'] = Financial_df.apply(lambda row: Separate_period_new_period(row), axis=1)
    Financial_df['Date'] = Financial_df.apply(lambda row: Separate_period_new_Date(row), axis=1)
    Financial_df['Date'] = pd.to_datetime(Financial_df['Date'])
    del Financial_df['index']
    # Financial_df = Financial_df.rename(columns = {'index':"Period as of"},index = {"Period as of" : " "})

    return Financial_df

def get_Financial_data(driver,stock,all_stock):
    
    column_name = {'index':"Period as of", 'Assets':"Assets",'Liabilities':"Liabilities",'Equity':"Equity",'Paid-up Capital':"Paid-up Capital",'Revenue':"Revenue",'Profit (Loss) from Other Activities':"Profit (Loss) from Other Activities",'Net Profit':"Net Profit",'EPS (Baht)':"EPS (Baht)",'ROA (%)':"ROA (%)",'ROE (%)':"ROE (%)",'Net Profit Margin (%)':"Net Profit Margin (%)",'Symbol':"Symbol"}
    stock = stock.split()[0]
    url_target_financial = "https://www.set.or.th/en/market/product/stock/quote/" + stock + "/financial-statement/company-highlights"
    driver.get(url_target_financial)
    
    Financial_data = driver.page_source
    Financial_df = pd.read_html(Financial_data)[0]
    
    Financial_df = operation_data(stock,Financial_df)
                   
    all_stock.append(Financial_df)
    print(stock)
    return driver , all_stock
```

```python
def get_set50_100_df(driver):
    
    data = driver.page_source
    data_df = pd.read_html(data)[2]
    clean_column = []
    for column in data_df.columns:

        clean_column.append(column.replace("  (Click to sort Ascending)",""))
        
    data_df.columns = clean_column
    data_df = data_df.set_index('Symbol')
    return data_df

def operation_data(stock,dataframe):
    
    Financial_df = (dataframe.set_index("งวดงบการเงิน ณ วันที่")).T
    Financial_df = Financial_df.drop(["บัญชีทางการเงินที่สำคัญ","อัตราส่วนทางการเงินที่สำคัญ"], axis=1)

    Financial_df['Symbol'] = stock
    Financial_df = Financial_df.reset_index()
    Financial_df = Financial_df.rename(columns = {'index':"งวดงบการเงิน ณ วันที่"},index = {"Period as of" : " "})

    return Financial_df

def get_Financial_data(driver,stock,all_stock):
    
    column_name = {'index':"งวดงบการเงิน ณ วันที่"}
    stock = stock.split()[0]
    url_target_financial = "https://www.set.or.th/th/market/product/stock/quote/" + stock + "/financial-statement/company-highlights"
    driver.get(url_target_financial)

    driver.execute_script("scrollBy(0,+5000);")
    time.sleep(3)
    
    Financial_data = driver.page_source
    Financial_df = pd.read_html(Financial_data)[0]
    
    Financial_df = operation_data(stock,Financial_df)
                   
    all_stock.append(Financial_df)
    print(stock)
    return driver , all_stock
```
