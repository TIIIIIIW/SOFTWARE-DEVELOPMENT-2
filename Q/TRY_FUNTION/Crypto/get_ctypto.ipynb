```python
def get_crypto(driver,p):

    driver.get("https://coinmarketcap.com/th/?page="+f'{p}')
    for i in range(18):
        driver.execute_script("scrollBy(0,+500);")
        time.sleep(1)

    df = pd.read_html(driver.page_source)[0]
    
    df = Crypto_clean(df,p)

    
    return driver,df

def Crypto_clean(df,p):

    Symbol,Name = [],[]
    round = (len(df['ชื่อ'])*p)

    n = 0
    for i in range(round-int(round/p),round):

        t = df['ชื่อ'][n].split(f'{i+1}')
        print(t)
        n+=1
        Symbol.append(t[1])
        Name.append(t[0])

    table = {}
    table['Symbol'] = Symbol
    table['Name'] = Name
    table['Market'] = 'CRYPTO'
    df = pd.DataFrame(table)

    return df
```

```python
all_t = []

for i in range(1,4):
    
    driver,table = get_crypto(driver,i)
    all_t.append(table)
```

```python
allCrypto = pd.concat(all_t)
```

```python
allCrypto.to_csv('Crypto300.csv',index=False)
```
