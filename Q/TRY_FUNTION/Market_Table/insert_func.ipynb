```python
def insert_market_share(data,market):

    conn = sqlite3.connect(r'C:\Users\JourneyQ\OneDrive - kmutnb.ac.th\Desktop\Quick_file\year_2-S_2\Project_sofedev_2\Manage_Share\Database_convert\share_V2.sqlite')
    cur = conn.cursor()

    check_query = f"""SELECT count(*),SymbolId FROM Information WHERE Symbol = "{data['Symbol']}" and Sname = "{data['Name']}";"""
    cur.execute(check_query)
    result_1 = cur.fetchone()

    if result_1[0] == 0:
        conn.close()
        return "0",data['Symbol']
        
    check_query = f"""SELECT count(*),I.SymbolId FROM Information as I INNER JOIN Sale as Sl on I.SymbolId = Sl.SymbolId INNER JOIN Market as M on M.MarketId = Sl.MarketId WHERE I.Symbol = "{data['Symbol']}" and I.Sname = "{data['Name']}" and M.Mname = '{data['Market']}';"""
    cur.execute(check_query)
    result_2 = cur.fetchone()

    if result_2[0] == 0:
        # Data does not exist, insert it
        insert_data = f""" INSERT INTO Sale("SymbolId","MarketId") VALUES("{result_1[1]}","{market[data['Market']]}")"""
        cur.execute(insert_data)
        conn.commit()
        conn.close()
        return "1",data['Symbol']
        
    conn.close()
    return "0",data['Symbol']
```
