```python
def getData(tickers) :
    data = []
    if tickers == [] :
        return 'No stock name'
    for ticker in tickers:
        detail = yf.download(ticker, interval='1D', start='2022-11-01')
        # detail = yf.download(ticker, interval='1h', start='2022-11-01', end='2022-12-01')
        detail['Name'] = ticker[:-3]
        table_price = detail.reset_index()
        data.append(table_price)
    return data
    

def putDataToCsv(datas):
    with open('WWA.csv', 'w') as creating_new_csv_file: 
        pass 
    count = 0
    for data in datas:
        if count > 0:
            data.to_csv('WWA.csv', mode='a', header=False,index=False)
            count+=1
        else:
            data.to_csv('WWA.csv', mode='a',index=False)
            count+=1
```
